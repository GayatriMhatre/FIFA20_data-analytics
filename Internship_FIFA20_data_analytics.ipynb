## Task 4. Data Preprocessing - Part 2

### Create DataFrame of each field positions and add prefix as (Position_)

### Concate above DataFrame with our original DataFrame

```python
new_player_position = data['player_positions'].str.get_dummies(sep=',').add_prefix('Position_')
new_player_position.head()
```

```python
data =  pd.concat([data,new_player_position],axis = 1)
```

```python
data.head()
```

## Task 5. Data Preprocessing - Part 3

### Clean Position Ratings Column

### Rating Columns - ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram','lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb','lcb', 'cb', 'rcb', 'rb']

### Fill Empty values = 0

### Make Data-type as = Integer

### Fill Missing Values of Skills

### Skills - ["dribbling", "defending", "physic", "passing", "shooting", "pace"]

### Fill Missing Values = As Median

```python
data =  data.drop('player_positions',axis=1)
```

```python
columns = ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram','lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb','lcb', 'cb', 'rcb', 'rb']
```

```python
data[columns].head()
```

```python
for col in columns:
  data[col]=data[col].str.split('+',n=1,expand = True)[0]

data[columns]
```

```python
data[columns] = data[columns].fillna(0)
```

```python
data[columns] = data[columns].astype(int)
```

```python
data[columns]
```

```python
columns = ['dribbling','defending','physic','passing','shooting','pace']
```

```python
data[columns]
```

```python
data[columns].isnull().sum()
```

```python
for col in columns:
  data[col] = data[col].fillna(data[col].median())
data[columns]
```

```python
data = data.fillna(0)
```

```python
data.isnull().sum()
```

## Task 6. Create Scatter Plot for (Overall Rating vs Value in Euros) [colored by Age]

```python
sns.relplot(x='overall',y='value_eur',hue='age',palette = 'viridis',size="BMI", sizes=(15, 200),aspect=2,data=data)
plt.title('Overall Rating vs Value in Euros',fontsize = 20)
plt.xlabel('Overall Rating')
plt.ylabel('Value in Euros')
plt.show()
```

## Task 8. Create Histogram of Players Ages

```python
plt.figure(dpi=125)
sns.distplot(a=data['age'],kde=False,bins=20,color='green')
plt.axvline(x=np.mean(data['age']),c='green',ls='--',label='Mean Age of All Players')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()
```

## Task 9. Describing the Percentage of Players in different (Attacker) positions

### attacker = ['RW', 'LW', 'ST', 'CF', 'LS', 'RS', 'RF', 'LF']

## Task 12. Top 10 players based on the player position and the player value in euro

```python
def pick_top_players(pos, value):
    column = str('Position_') + str.upper(pos)
    if column in list(data.columns):
        target_players = data[(data[column] == 1) & (data['value_eur'] <= value)][['short_name', 'age', 'overall', 'value_eur']].head(10)
        return target_players
    else:
        print('No players in this position')
```
